# Small Sample Tests

## One Sample t- test

### Two sided test

```python
df_production = pd.read_csv(r"D:\Datasets\Production.csv")
df_production
```

Let <br>X: the number of furniture items produced by machine A <br>
Thus $\mu$: the average number of furniture items produced by machine A <br>

The hypothesis is                                                                       
H<sub>0</sub>: The mean number of items produced my machine A is equal to 30, i.e $\mu$= 30 <br>
**AGAINST**<br>
H<sub>a</sub>: The mean number of items produced my machine A is not equal to 30, , i.e $\mu$ ≠ 30

```python
# load the data
df_production = pd.read_csv(r"D:\Datasets\Production.csv")

# t-test for one sample
from scipy.stats import ttest_1samp

# conduct the test
ttest_1samp(df_production["Items Produced"],30)
```

We see that the p-value (0.236) is greater than 0.05 (level of significance), hence we accept H<sub>0</sub>.<br> We may conclude that the average number furniture items produced by machine A is equal to 30

# One sided test 

```python
df_lifeSpan = pd.read_csv(r"D:\Datasets\LifeSpan.csv")
df_lifeSpan
```

Let <br>X: the life of bulb in hours <br>
Thus $\mu$<sub>A</sub>: the average life of bulb in hours 

The hypothesis is                                                                     
H<sub>0</sub>: The mean life of bulb in hours is equal to 45000, i.e $\mu$<sub>A</sub> = 45000 <br>
**AGAINST**<br>
H<sub>a</sub>: The mean life of bulb in hours is less than or equal to 45000, i.e $\mu$<sub>A</sub> ≤ 45000

```python
# load the data 
df_lifeSpan = pd.read_csv(r"D:\Datasets\LifeSpan.csv")

# conduct the test
# this is the test is two tailed
t_test_less = ttest_1samp(df_lifeSpan["life"], 45000)

# to get the p-value for one tailed test
# divide the p-value by 2
p_value_less = t_test_less.pvalue/2

# print p-value
p_value_less
```

We see that the p-value (0.02) is less than 0.05 (level of significance), hence we reject H<sub>0</sub>. <br>We may conclude that the average life of bulb in hours is less than or equal to 45000. And the manufacturers claim is false.

## Two Sample t- test

```python
df_diet = pd.read_csv(r"D:\Datasets\Diet_data.csv")
df_diet
```

Let <br>X: the gain in weight due to diet A <br>
    Y: the gain in weight due to diet B<br><br>
Thus <br>
$\mu$<sub>A</sub>: the average gain in weight due to diet A <br>
$\mu$<sub>B</sub>: the average gain in weight due to diet B <br>
    
                                                                  
The hypothesis is                                                                     
H<sub>0</sub>: The average gain in weight due to diet A is equal to the average gain in weight due to diet B, i.e $\mu$<sub>A</sub> = $\mu$<sub>B</sub> <br>
**AGAINST**<br>
H<sub>a</sub>: The average gain in weight due to diet A is not equal to the average gain in weight due to diet B, i.e $\mu$<sub>A</sub> ≠ $\mu$<sub>B</sub> <br>

```python
# load the data 
df_diet = pd.read_csv(r"D:\Datasets\Diet_data.csv")

# t-test for testing equality of means
from scipy.stats import ttest_ind

# conduct the test
ttest_ind(df_diet["Diet A"],df_diet["Diet B"])
```

Since the p-value (0.611) is more than 0.05 (level of significance), hence we accept H<sub>0</sub>. <br> We may conclude that the average gain in weight due to diet A and diet B are not the same.

## Paired t test

```python
df_Sleep = pd.read_csv(r"D:\Datasets\Sleep_data.csv")
df_Sleep
```

Let <br>X: hours of sleep before medication <br>
    Y: hours of sleep after medication<br><br>

$\mu$<sub>A</sub>: the average hours of sleep before medication <br>
$\mu$<sub>B</sub>: the average hours of sleep after medication <br>

# Normality

### Shapiro test

```python
# create the data from by genrating a random sample
rainfall_in_mm = np.random.random(200)*100
```

The null and alternate hypothesis of Shapiro-Wilk test are as follows: <br>
    
H<sub>0</sub>: The data is normally distributed<br>
**AGAINST**<br>
H<sub>a</sub>: The data is not normally distributed

```python
# import shapiro from the scipy library
from scipy.stats import shapiro

# conduct the normality test
stat, p = shapiro(rainfall_in_mm)

print("The test statistic is", stat," and its p-value is",p)
```

Since the p-value (0.00) is less than 0.05 (level of significance), hence we reject H<sub>0</sub>. <br> We may conclude that the data is not normally distributed.

### Jarque-Bera test

```python
# create the data from by genrating a random sample
diameter_in_mm = np.random.normal(loc=4, scale =0.7, size=6000) + np.random.normal(loc=5, scale =0.1, size=6000)
```

The null and alternate hypothesis of Jarque-Bera test are as follows: <br>
    
H<sub>0</sub>: The data is normally distributed<br>
**AGAINST**<br>
H<sub>a</sub>: The data is not normally distributed

```python
# import shapiro from the scipy library
from scipy.stats import jarque_bera

# conduct the normality test
stat, p = jarque_bera(diameter_in_mm)

print("The test statistic is", stat," and its p-value is",p)
```

Since the p-value (0.125) is more than 0.05 (level of significance), hence we accept H<sub>0</sub>. <br> We may conclude that the data is normally distributed.
